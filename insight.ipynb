```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np


folder_path = "./100_case_si/processed/"
# folder_path = "./130_case_test/"
max_values = []


csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    if 'p_convert' in df.columns:
        max_p = df['p_convert'].max()
        max_values.append(max_p)
        
space = 1

bin_edges = np.arange(0, max(max_values) + space, space)

# Tính tần suất xuất hiện trong mỗi bin
hist, bins = np.histogram(max_values, bins=bin_edges)

plt.figure(figsize=(12, 6))
plt.bar(bins[:-1], hist, width=space, align="edge", color="salmon", edgecolor="black")
plt.xlabel("Range of max p_convert Values")
plt.ylabel("Frequency")
plt.title(f"Frequency Distribution of Max p_convert Values (bin size = {space})")
plt.xticks(bins, rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
